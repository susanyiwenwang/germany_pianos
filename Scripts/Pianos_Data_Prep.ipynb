## Normalize Trade Values for Inflation

```python
# import inflation dataset
infl = pd.read_csv(os.path.join(path, 'Data', 'us_cpi_u.csv'), index_col=False)
```

```python
infl
```

```python
# merge with main df
df1 = df.merge(infl, how='left', on='Year')
df1.head()
```

```python
df1.tail()
```

```python
# create new columns for all trade values with inflation calcuation for year 2024
# price_2024 = price_year * (313.7/ cpi_year)
df1['Upright Import Trade Value'] = (df1['Upright Import Trade Value 1000USD']*(313.7/ df1['CPI_U'])).round(2)
df1['Grand Import Trade Value'] = (df1['Grand Import Trade Value 1000USD']*(313.7/ df1['CPI_U'])).round(2)
df1['Keyboard Import Trade Value'] = (df1['Keyboard Import Trade Value 1000USD']*(313.7/ df1['CPI_U'])).round(2)

df1['Upright Export Trade Value'] = (df1['Upright Export Trade Value 1000USD']*(313.7/ df1['CPI_U'])).round(2)
df1['Grand Export Trade Value'] = (df1['Grand Export Trade Value 1000USD']*(313.7/ df1['CPI_U'])).round(2)
df1['Keyboard Export Trade Value'] = (df1['Keyboard Export Trade Value 1000USD']*(313.7/ df1['CPI_U'])).round(2)
```

```python
df1.head()
```

```python
# remove original trade value columns
keep = list()
for col in df1.columns:
    if '1000USD' not in col:
        keep.append(col)
keep.remove('CPI_U')
print(keep)
```

```python
df2 = df1[keep].copy()
df2.head()
```

## Price per instrument columns

```python
df2['Upright Import Price'] = df2['Upright Import Trade Value']/df2['Upright Import Quantity']
df2['Grand Import Price'] = df2['Grand Import Trade Value']/df2['Grand Import Quantity']
df2['Keyboard Import Price'] = df2['Keyboard Import Trade Value']/df2['Keyboard Import Quantity']

df2['Upright Export Price'] = df2['Upright Export Trade Value']/df2['Upright Export Quantity']
df2['Grand Export Price'] = df2['Grand Export Trade Value']/df2['Grand Export Quantity']
df2['Keyboard Export Price'] = df2['Keyboard Export Trade Value']/df2['Keyboard Export Quantity']
```
